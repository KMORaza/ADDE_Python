```python
import numpy as np
A = np.array([[3, 0, -1],
              [0, 1, 0],
              [2, 0, 0]])
eigenvalues_A = np.linalg.eigvals(A)
print("Eigenvalues of A =", eigenvalues_A)
eigenvalues_A, eigenvectors_A = np.linalg.eig(A)
print("Eigenvalues of A =", eigenvalues_A)
print("Eigenvectors of A =")
print(eigenvectors_A)
p = np.poly(A)
print("Coefficients of characteristic polynomial =", p)
r = np.roots(p)
print("Roots of characteristic polynomial =", r)
sum_of_eigenvalues = np.trace(A)
print("Sum of eigenvalues =", sum_of_eigenvalues)
product_of_eigenvalues = np.prod(r)
print("Product of eigenvalues =", product_of_eigenvalues)
det_A = np.linalg.det(A)
print("Determinant of A =", det_A)
inv_A = np.linalg.inv(A)
print("Inverse of A =")
print(inv_A)
eigenvalues_inv_A = np.linalg.eigvals(inv_A)
print("Eigenvalues of inverse of A =", eigenvalues_inv_A)
eigenvalues_transpose_A = np.linalg.eigvals(A.T)
print("Eigenvalues of transpose of A =", eigenvalues_transpose_A)
B = np.linalg.eigvals(A**2 + 3*A + 2*np.eye(3))
print("Eigenvalues of B =", B)
```

```python
import numpy as np
import matplotlib.pyplot as plt
c = 0.5
P = 1
L = 5
x = np.arange(0, L + 0.1, 0.1)
y = np.arange(-c, c + 0.01, 0.01)
X, Y = np.meshgrid(x, y)
sx = -(3 / (4 * c**3)) * (L - X) * Y
sy = np.zeros_like(sx)
txy = -(3 / (c**3)) * (c**2 - Y**2)
s1 = np.zeros_like(sx)
s2 = np.zeros_like(sx)
for i in range(len(y)):
    for j in range(len(x)):
        s = np.array([[sx[i, j], txy[i, j]],
                      [txy[i, j], sy[i, j]]])
        p = np.linalg.eigvals(s)
        s1[i, j] = p[1]
        s2[i, j] = p[0]
plt.figure(1)
contour_levels_s1 = [0.01, 0.05, 0.1, 0.5, 1, 3, 5, 7, 9, 11]
plt.contour(X, Y, s1, levels=contour_levels_s1)
plt.title('First Principle Stress Contours, s1')
plt.axis('tight')
plt.figure(2)
contour_levels_s2 = [-11, -9, -7, -5, -3, -1, -0.5, -0.1, -0.05, -0.01]
plt.contour(X, Y, s2, levels=contour_levels_s2)
plt.title('Second Principle Stress Contours, s2')
plt.axis('tight')
plt.show()
```

**`Experiment-2`**

```python
import numpy as np
def pagerank(M, num_iterations=100, d=0.85):
    N = M.shape[0]
    r = np.ones(N) / N
    for _ in range(num_iterations):
        r = d * M @ r + (1 - d) / N * np.ones(N)
    return r
link_matrix = np.array([[0, 1, 0],
                        [1, 0, 1],
                        [1, 1, 0]])
M = link_matrix / np.sum(link_matrix, axis=1, keepdims=True)
print("Link Matrix:")
print(link_matrix)
print("\nTransition Matrix (Stochastic Matrix):")
print(M)
eigenvalues_M = np.linalg.eigvals(M)
print("\nEigenvalues of Transition Matrix:")
print(eigenvalues_M)
trace_M = np.trace(M)
print("\nTrace of Transition Matrix:")
print(trace_M)
try:
    inverse_M = np.linalg.inv(M)
    print("\nInverse of Transition Matrix:")
    print(inverse_M)
except np.linalg.LinAlgError:
    print("\nError: Transition Matrix is singular. It does not have a unique inverse.")
transpose_M = M.T
print("\nTranspose of Transition Matrix:")
print(transpose_M)
out_degree = np.sum(link_matrix, axis=1)
print("\nOut-Degree of Each Page:")
print(out_degree)
in_degree = np.sum(link_matrix, axis=0)
print("\nIn-Degree of Each Page:")
print(in_degree)
page_rank_scores = pagerank(M)
print("\nPageRank Scores:", page_rank_scores)
```

```python
import sympy as sp
t = sp.symbols('t')
x1 = sp.Function('x1')(t)
x2 = sp.Function('x2')(t)
A = sp.Matrix([[1, 2], [3, 4]])
lambda_A = A.eigenvals()
print("Eigenvalues of A =", lambda_A)
P = sp.zeros(A.shape[0], A.shape[0])
for i, eigenvalue in enumerate(lambda_A.keys()):
    temp = (A - eigenvalue * sp.eye(A.shape[0])).nullspace()[0]
    P[:, i] = temp / min(temp)
print("\nModel Matrix P =",P)
print(P)
D = P.inv() * A * P
eq1 = sp.diff(x1, t, 2) + D[0] * x1
eq2 = sp.diff(x2, t, 2) + D[3] * x2
Sol1 = sp.dsolve(eq1)
Sol2 = sp.dsolve(eq2)
Y = P * sp.Matrix([Sol1.rhs, Sol2.rhs])
print("\nSolution for x1(t) =", Sol1)
print("\nSolution for x2(t) =", Sol2)
print("\nCombined solution Y =", Y)
Y = P * sp.Matrix([Sol1.rhs, Sol2.rhs])
print("\nOutput:")
print("Y =", Y)
print(f"x1(t) = {Sol1.rhs}")
print(f"x2(t) = {Sol2.rhs}")
print(f"Combined solution Y = {Y}")
```

```python
from sympy import symbols, laplace_transform, inverse_laplace_transform, Function, Eq, dsolve, sin, cos
t, s = symbols('t s')
Y = Function('Y')(s)
ode = Eq(s**2*Y + 4*s*Y + 4*Y, 3/(s**2 + 1) + 4*s/(s**2 + 1))
Y_solutions = dsolve(ode, Y)
Y_particular = Y_solutions.rhs
y_t = inverse_laplace_transform(Y_particular, s, t)
print("Solution:")
print("y(t) =", y_t.simplify())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
L = 2.0
E = 2.1e11
I = 4.5e-11
def w(x):
    if 0 < x < 1:
        return x
    elif 1 <= x <= L:
        return x - L
    else:
        return 0
def deflection_ode(x, y):
    return [y[1], y[2], y[3], w(x) / (E * I)]
initial_conditions = [0, 0, 0, 0]
solution_part1 = solve_ivp(deflection_ode, [0, 1], initial_conditions, t_eval=np.linspace(0, 1, 100))
initial_conditions_part2 = solution_part1.y[:, -1]
solution_part2 = solve_ivp(deflection_ode, [1, L], initial_conditions_part2, t_eval=np.linspace(1, L, 100))
x_values = np.concatenate([solution_part1.t, solution_part2.t])
y_values = np.concatenate([solution_part1.y[0], solution_part2.y[0]])
plt.plot(x_values, y_values, label='Deflection')
plt.title('Cantilever Beam Deflection')
plt.xlabel('Position along the beam (mm)')
plt.ylabel('Deflection (mm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
x = sp.symbols('x')
def f(x):
    return sp.sin(x)
def calculate_fourier_coefficients(f, a, b, N):
    coefficients = []
    L = b - a
    for n in range(N + 1):
        an = (2 / L) * sp.integrate(f(x) * sp.cos(2 * sp.pi * n * (x - a) / L), (x, a, b))
        bn = (2 / L) * sp.integrate(f(x) * sp.sin(2 * sp.pi * n * (x - a) / L), (x, a, b))
        coefficients.append((an, bn))
    return coefficients
def fourier_series(x, a, b, coefficients):
    L = b - a
    result = 0.5 * coefficients[0][0]
    N = len(coefficients) - 1
    for n in range(1, N + 1):
        result += coefficients[n][0] * sp.cos(2 * sp.pi * n * (x - a) / L) + coefficients[n][1] * sp.sin(2 * sp.pi * n * (x - a) / L)
    return result
a = 0
b = 2 * sp.pi
N = 5
x_values = np.linspace(float(a), float(b), 1000)
coefficients = calculate_fourier_coefficients(f, a, b, N)
y_values = [fourier_series(x, a, b, coefficients).subs('x', val).evalf() for val in x_values]
plt.figure(figsize=(10, 6))
plt.plot(x_values, [f(val) for val in x_values], label='Original Function')
plt.plot(x_values, y_values, label=f'Fourier Series (N={N})', linestyle='--')
plt.title('Fourier Series')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
